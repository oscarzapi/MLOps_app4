As this statistical information shows, this variable has large difference between minimum and maximum values, and the mean value is far from the cheapest and most expensive properties, so there may be unusual values.

To check this fact, a boxplot is shown below:

```python
sns.boxplot(data.price, color="y")
```

Visually, it is easier to see there are certain values that are over the maximum value established by the boxplot diagram:
\begin{equation*}
Ls = Q_{3} + 1.5R.I.C
\end{equation*}

\begin{equation*}
R.I.C = Q_{3} - Q_{1}
\end{equation*}

Where $Q_{3}$ is the third quartile and $Q_{1}$ is the first quartile. In this case, $R.I.C = 1719900 - 35000 = 1684900$ and $Ls = 1719900 + 1.5(1684900) = 4247250$

So, all values for 'price' over $Ls$ will be considered as outliers and therefore their removal will be studied.

##  Outliers

Outlier analysis will be based over the most important variables. A dataframe will be created to select the observations to be deleted from the original dataset.

```python
train_outliers = pd.DataFrame()
```

```python
train_outliers['price'] = data['price'].copy()
```

### Outliers of 'price' variable

We will delete those properties whose values are over $Ls$ and the ones with value lower than 1, because we do not think someone wants to sell the property for free or we can have any problems when working with zero values.

```python
# outliers detection
upper_range = 4247250

# observations to be deleted
print(" Count of outliers over Ls: " + str(train_outliers["price"][train_outliers["price"]>upper_range].count()))
print(" Count of outliers less than 1: " + str(train_outliers["price"][train_outliers["price"]<1].count()))
```

```python
# outliers removal in original data
data = data.drop(data[train_outliers["price"]>upper_range].index)
data = data.drop(data[train_outliers["price"]<1].index)


# outliers removal on train_outliers dataset
train_outliers = train_outliers.drop(train_outliers[train_outliers["price"]>upper_range].index)
train_outliers = train_outliers.drop(train_outliers[train_outliers["price"]<1].index)
```

```python
# data logarithm
train_outliers['price'] = np.log(train_outliers['price'])
```

```python
# Comparison between original chart over chart with logarithm applied.
fig, axs = plt.subplots(1,2)
sns.distplot(data['price'], fit=norm, ax=axs[0])
sns.distplot(train_outliers['price'], fit=norm, ax=axs[1], axlabel = "price logarithm")
```

## 7. Calculate the average duration it takes to close a post and mention the parameters that affect this duration. Also, mention the district with the shortest duration it takes to close a post with property type of villa. <a id='section_7'></a>

The average time it takes to close a post could be done with the following expression:

```python
sales_dataset['Brother_Update_Time'] = pd.to_datetime(sales_dataset['Brother_Update_Time'])
sales_dataset['Creation_Time'] = pd.to_datetime(sales_dataset['Creation_Time'])

sales_dataset['post_duration'] = sales_dataset["Brother_Update_Time"] - sales_dataset["Creation_Time"]
```

```python
avg_duration = sales_dataset[(sales_dataset["Closed"] == "مغلق")]["post_duration"].mean()
```

```python
print("So for closed posts, the average time to close them is: " + str(avg_duration))
```

The main parameters involved for this duration are:
- "Brother_Update_Time" is the updated time for that post.
- "Creation_Time" is the time when the post was created.

Now to solve the second point, let's compute this:

```python
villa_avg_duration = sales_dataset[(sales_dataset["Property_Type"] == "فيلا") # Setting condition of 'villa' property
                                   & (sales_dataset["Closed"] == "مغلق") # Setting condition of 'Closed' status
                                   & (sales_dataset["Brother_Update_Time"] != sales_dataset["Creation_Time"]) # Setting condition of not taking into account equal values for updating and creating posts. It does not make sense that posts are updated inmmediately.
                                  ]["post_duration"].groupby(sales_dataset["district_name_en"]).min().sort_values(ascending = True)
```

```python
villa_avg_duration
```

The previous table shows that the district called 'Hittin' has the shortest interval of time to close a post with 9 seconds. In contrast, At Tarif is not a preferable district to post a property because you may get bored way before it is sold.
